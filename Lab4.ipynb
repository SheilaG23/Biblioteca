```python
import pandas as pd
```

```python
file_path= ("CSV Laboratorio 4 AD-B.csv")
data=pd.read_csv(file_path)
data.head(10)
```

```python
rows, columns = data.shape
print(f"El dataset tiene {rows} filas y {columns} columnas.")
```

```python
data.info()
data.describe()
```

```python
missing_values = data.isnull().sum()
print(missing_values)
```

```python
duplicated_rows = data.duplicated().sum()
print(f"Duplicados: {duplicated_rows}")
```

```python
data_cleaned=data.drop_duplicates()
```

```python
data["Category"]=data["Category"].fillna("Unknown")
```

```python
city_value_counts = data["City"].value_counts()
print(city_value_counts)
```

```python
data["City"]=data["City"].fillna("Unknown city")
```

```python
if 'Value' in data.columns:
    data['Value'] = data['Value'].fillna(data['Value'].median())
```

```python
if 'City' in data.columns:
 data['City'] = data['City'].str.lower()
```

```python
if 'Date' in data.columns:
 data['Date'] = pd.to_datetime(data['Date'], errors='coerce').dt.strftime('%Y/%m/%d')
```
